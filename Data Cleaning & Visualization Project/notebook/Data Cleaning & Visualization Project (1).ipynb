Key Insights from Your Dataset

Dataset Size:

500 student records
11 columns

Major Insights

1. Pass Percentage Analysis
Around 64.6% students passed
Around 35.4% students failed
Insight:

Most students are successfully passing, but a significant percentage still struggles academically.

2. Study Hours Strongly Affect Final Scores

There is a strong positive correlation between:

Study Hours Per Week
Final Score

Correlation value ≈ 0.80

Insight:

Students who study more hours generally achieve better academic performance.

This is one of the strongest findings in your project.

3. Attendance Has Moderate Impact

Attendance rate has a moderate positive relationship with scores.

Insight:

Regular attendance improves student performance, though not as strongly as study hours.

4. Gender-Based Performance

Average Final Scores:

Female ≈ 55.5
Male ≈ 56.5
Insight:

Academic performance is nearly balanced between genders.

5. Internet Access Helps Students

Students with internet access scored slightly higher.

Insight:

Digital resources and online learning positively influence academic performance.

6. Extracurricular Activities

Students participating in extracurricular activities performed slightly better.

Insight:

Balanced academic and extracurricular involvement may improve student engagement and outcomes.

7. Parent Education Influence

Students whose parents had higher education levels generally scored better.

Highest average scores came from:

PhD
Master’s degree holders
Insight:

Parental educational background may positively influence student learning environments.
