# Exploratory Data Analysis on Student Academic Performance Dataset

## Objective
The objective of this project is to analyze student academic performance data to uncover hidden patterns, identify influencing factors, and understand trends affecting student outcomes using statistical summaries and visualizations.

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (8,5)
```

```python
# Load Dataset
df = pd.read_csv("cleaned_student_data.csv")
```

```python
#Gender Distribution
sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
#Pass vs Fail Analysis
sns.countplot(x='passed', data=df)
plt.title("Pass vs Fail Distribution")
plt.show()
```

```python
#Final Score Distribution
sns.histplot(df['final_score'], kde=True)

plt.title("Distribution of Final Scores")
plt.show()
```

```python
#Study Hours vs Final Score
sns.scatterplot(
    x='study_hours_per_week',
    y='final_score',
    data=df
)

plt.title("Study Hours vs Final Score")
plt.show()
```

```python
#Attendance vs Final Score
sns.scatterplot(
    x='attendance_rate',
    y='final_score',
    data=df
)

plt.title("Attendance Rate vs Final Score")
plt.show()
```

```python
#Previous Score vs Final Score
sns.scatterplot(
    x='previous_score',
    y='final_score',
    data=df
)

plt.title("Previous Score vs Final Score")
plt.show()
```

```python
#Parent Education Level Analysis
sns.barplot(
    x='parent_education',
    y='final_score',
    data=df
)

plt.xticks(rotation=45)

plt.title("Parent Education Level vs Final Score")
plt.show()
```

```python
#Internet Access Impact
sns.boxplot(
    x='internet_access',
    y='final_score',
    data=df
)

plt.title("Internet Access vs Final Score")
plt.show()
```

```python
#Extracurricular Activities
sns.boxplot(
    x='extracurricular',
    y='final_score',
    data=df
)

plt.title("Extracurricular Activities vs Final Score")
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(12,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
#Pairplot
sns.pairplot(df)
plt.show()
```

```python
#Outlier Detection
sns.boxplot(x=df['final_score'])

plt.title("Outlier Detection in Final Scores")
plt.show()
```

# Key Findings

1. Students who studied more hours generally achieved better final scores.

2. Attendance rate positively influenced academic performance.

3. Previous academic performance strongly correlated with final scores.

4. Internet access slightly improved student performance.

5. Parent education level had noticeable influence on academic success.

6. Participation in extracurricular activities showed moderate performance benefits.

7. Gender differences in performance were minimal.

# Conclusion

This Exploratory Data Analysis project successfully identified the major factors affecting student academic performance. Study habits, attendance, previous academic scores, and access to educational resources significantly influenced student outcomes.

The analysis demonstrated how EDA techniques help uncover meaningful insights and trends from educational datasets.
